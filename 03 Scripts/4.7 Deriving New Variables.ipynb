## 3. If -statements

```python
## Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## 5. If-Statements with For-Loops

```python
df_ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

#### Legende: Variables  order_day_of_week 
     
      0 = Saturday            
      1 = Sunday               
      2 = Monday               
      3 = Tuesday
      4 = Wednesday
      5 = Thursday
      6 = Friday

```python
result = []

for value in df_ords_prods_merged["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged['Busiest day'] = result
```

```python
df_ords_prods_merged['Busiest day'].value_counts(dropna = False)
```

## 4.7 task - Deriving New Variable

#### - Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
result = []
for value in df_ords_prods_merged["order_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 3 or value == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")

df_ords_prods_merged["Busyness_Level"] = result
```

```python
result
```

```python
df_ords_prods_merged['Busiest days'] = result
```

```python
df_ords_prods_merged['Busiest days'].value_counts(dropna = False)
```

```python
## Find out the 25 %, 50% and 75% for If statement 'order_hour_of_day' column 

df_ords_prods_merged['order_hour_of_day'].describe()
```

```python
## Now 25 % are 10 orders, 50% are 13 orders and 75 % are 16 ordes, so the function is as follow:

result = []
for value in df_ords_prods_merged["order_hour_of_day"]:
    if value >= 16:
        result.append("Most orders")
    elif value > 10 and value <= 14:
        result.append("Average orders")
    else:
        result.append("Fewest orders")

df_ords_prods_merged["busiest_period_of_day"] = result
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result
```
